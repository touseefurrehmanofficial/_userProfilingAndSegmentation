## Segment Users into Groups for Targeted Ad Campaigns

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline

# Convert age ranges to their midpoints
def convert_age_range(age_range):
    try:
        age_range = str(age_range).strip()
        if '-' in age_range:
            start, end = age_range.split('-')
            return (int(start) + int(end)) / 2
        elif age_range.endswith('+'):
            return int(age_range[:-1]) + 2 
        else:
            return float(age_range)
    except ValueError:
        return np.nan

data['Age'] = data['Age'].apply(convert_age_range)
```

```python
# Ensure column names
data.columns = [col.strip() for col in data.columns]

# Select relevant features for clustering
features = ['Age', 'Gender', 'Income Level', 'Education Level', 'Time Spent Online (hrs/weekday)',
            'Time Spent Online (hrs/weekend)', 'Likes and Reactions', 'Click-Through Rates (CTR)', 'Conversion Rates', 'Top Interests']


preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['Age', 'Time Spent Online (hrs/weekday)', 'Time Spent Online (hrs/weekend)', 'Likes and Reactions', 'Click-Through Rates (CTR)', 'Conversion Rates']),
        ('cat', OneHotEncoder(), ['Gender', 'Income Level', 'Education Level', 'Top Interests'])
    ])

# Create a pipeline
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('kmeans', KMeans(n_clusters=5, random_state=42))
])

# Fit the model
pipeline.fit(data[features])
```

```python
# Add the cluster labels
data['Cluster'] = pipeline['kmeans'].labels_

print(data.head())
```

## Compute Mean Values of Numerical Features and Mode for Categorical Features within Each Cluster

```python
def mode(series):
    return series.mode().iloc[0]

# Compute mean values for numerical features and mode for categorical features
cluster_summary = data.groupby('Cluster').agg({
    'Age': 'mean',
    'Time Spent Online (hrs/weekday)': 'mean',
    'Time Spent Online (hrs/weekend)': 'mean',
    'Likes and Reactions': 'mean',
    'Click-Through Rates (CTR)': 'mean',
    'Conversion Rates': 'mean',
    'Ad Interaction Time (sec)': 'mean',
    'Gender': mode,
    'Location': mode,
    'Language': mode,
    'Education Level': mode,
    'Income Level': mode,
    'Top Interests': mode
}).reset_index()


print(cluster_summary)
```

```python
# Define cluster names
cluster_names = {
    0: "Weekend Warriors",
    1: "Engaged Professionals",
    2: "Low-Key Users",
    3: "Active Explorers",
    4: "Budget Browsers"
}

# Add cluster names
cluster_summary['Cluster Name'] = cluster_summary['Cluster'].map(cluster_names)

print(cluster_summary)
```

```python
for index, row in cluster_summary.iterrows():
    print(f"Cluster {row['Cluster']}  {row['Cluster Name']}:")
    print(f"  - Age: {row['Age']:.2f}")
    print(f"  - Time Spent Online (hrs/weekday): {row['Time Spent Online (hrs/weekday)']:.2f}")
    print(f"  - Time Spent Online (hrs/weekend): {row['Time Spent Online (hrs/weekend)']:.2f}")
    print(f"  - Likes and Reactions: {row['Likes and Reactions']:.2f}")
    print(f"  - Click-Through Rates (CTR): {row['Click-Through Rates (CTR)']:.3f}")
    print(f"  - Conversion Rates: {row['Conversion Rates']:.3f}")
    print(f"  - Ad Interaction Time (sec): {row['Ad Interaction Time (sec)']:.2f}")
    print(f"  - Gender: {row['Gender']}")
    print(f"  - Location: {row['Location']}")
    print(f"  - Language: {row['Language']}")
    print(f"  - Education Level: {row['Education Level']}")
    print(f"  - Income Level: {row['Income Level']}")
    print(f"  - Top Interests: {row['Top Interests']}")
    print("")
```

## Create a Radar Chart to Reflect Segments

```python
import plotly.express as px
import plotly.offline as pyo

cluster_summary_normalized = cluster_summary.copy()
features_to_normalize = ['Time Spent Online (hrs/weekday)', 'Time Spent Online (hrs/weekend)', 'Likes and Reactions', 'Click-Through Rates (CTR)']
for feature in features_to_normalize:
    cluster_summary_normalized[feature] = cluster_summary_normalized[feature] / cluster_summary_normalized[feature].max()

cluster_summary_melted = cluster_summary_normalized.melt(id_vars=['Cluster', 'Cluster Name'], value_vars=features_to_normalize, var_name='Feature', value_name='Value')

fig = px.line_polar(cluster_summary_melted, r='Value', theta='Feature', color='Cluster Name', line_close=True, title="User Segments Profile")
fig.update_traces(fill='toself')
fig.show()
```
